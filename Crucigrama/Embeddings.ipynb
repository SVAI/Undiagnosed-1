```
import pandas as pd
from glove import Corpus, Glove
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

```

```
df = pd.read_csv("vep.genes.tsv", sep="\t", header=None,names=['gene','num','name'])  
```

```
values = list(map(lambda x: x.split(' '),df['name']))
```

```
values
```

```
corpus = Corpus() 

corpus.fit(values, window=10)

glove = Glove(no_components=5, learning_rate=0.5)
 
glove.fit(corpus.matrix, epochs=10000, no_threads=8, verbose=False)
glove.add_dictionary(corpus.dictionary)
glove.save('glove.model')
```

```
glove.word_vectors
```

```
glove.dictionary

glove.idx_to_text = { glove.dictionary[key]: key for key in glove.dictionary.keys()}
```

```
X_embedded = TSNE(n_components=2).fit_transform(glove.word_vectors)
```
