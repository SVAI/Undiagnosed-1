```
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```
base_url = 'https://ghr.nlm.nih.gov/gene/'

def scrap_site(gene):
  response = requests.get(base_url + gene)
  soup = BeautifulSoup(response.text, "html.parser")
  
  descr = ""

  for div_tag in soup.findAll('div',{'class': 'expandBoxContent'}):
    descr += div_tag.text
    
  return descr.trim()
  
```

```
with open("vep.genes.tsv") as file:
  
  for line in file:
    gene = line.split('\t')[0]
    gene_data = scrap_site(gene)
    print(gene + '\t' + gene_data)
```
